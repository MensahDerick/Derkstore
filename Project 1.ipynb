# Statistical inferences
#### 1. Mean 
#### 2. Median
#### 3. Correlation 
#### 4. Standard deviation
#### 5. Variance 

```python
cols = ["fixed_acidity", "volatile_acidity", "citricAcid", "residual_sugar", "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density", "pH", "sulphates", "alcohol", "quality"]
```

```python
means = df[cols].mean()
```

```python
means
```

```python
df[cols].median()
```

```python
df[cols].corr()
```

```python
df[cols].std()
```

```python
df[cols].var()
```

```python
data = means
keys = ["fixed_acidity", "volatile_acidity", "citricAcid", "residual_sugar", "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density", "pH", "sulphates", "alcohol", "quality"]
palette_color = sb.color_palette('dark')
# plt.axis("equal")

plt.pie(data, labels = keys, colors = palette_color, autopct = "%20f%%" , startangle = 90, radius = 6)
plt.show()
```

```python
# sb.histplot(data = means, count = keys )
```

```python
# sb.barplot(data = means)
y = means
x = keys

plt.bar(x = x, height =  y, width = 0.7)
plt.xlabel("Keys")
plt.ylabel("Means")
plt.show()
```
